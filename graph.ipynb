```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

data = pd.read_csv('data.csv')
#select rows where country name is 'United States' or 'Euro area' or 'China'
data = data[(data['Country Name'] == 'United States') | (data['Country Name'] == 'European Union') | (data['Country Name'] == 'China') | (data['Country Name'] == 'Brazil') | (data['Country Name'] == 'World')]
data.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Drop column: '1960'
    data = data.drop(columns=['1960'])
    # Drop column: '2021'
    data = data.drop(columns=['2021'])
    # Drop column: '2022'
    data = data.drop(columns=['2022'])
    # Drop column: 'Unnamed: 67'
    return data

data_clean = clean_data(data.copy())
data_clean.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data_clean):
    # Drop column: 'Indicator Code'
    data_clean = data_clean.drop(columns=['Indicator Code'])
    return data_clean

data_clean_1 = clean_data(data_clean.copy())
data_clean_1.head()
```

```python
brazil = data_clean_1[data_clean_1['Country Name'] == 'Brazil']
china = data_clean_1[data_clean_1['Country Name'] == 'China']
euro = data_clean_1[data_clean_1['Country Name'] == 'European Union']
usa = data_clean_1[data_clean_1['Country Name'] == 'United States']
world = data_clean_1[data_clean_1['Country Name'] == 'World']

#remove indexs of countries
china = china.reset_index(drop=True)
euro = euro.reset_index(drop=True)
usa = usa.reset_index(drop=True)
brazil = brazil.reset_index(drop=True)
world = world.reset_index(drop=True)

brazil.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(brazil):
    # Drop column: 'Country Name'
    brazil = brazil.drop(columns=['Country Name'])
    # Drop column: 'Country Code'
    brazil = brazil.drop(columns=['Country Code'])
    # Drop column: 'Indicator Name'
    brazil = brazil.drop(columns=['Indicator Name'])
    return brazil

brazil_clean = clean_data(brazil.copy())
china_clean = clean_data(china.copy())
euro_clean = clean_data(euro.copy())
usa_clean = clean_data(usa.copy())
world_clean = clean_data(world.copy())
```

```python
usa_values = np.array(usa_clean)
china_values = np.array(china_clean)
euro_values = np.array(euro_clean)
brazil_values = np.array(brazil_clean)
world_values = np.array(world_clean)
usa_values = usa_values.flatten()
china_values = china_values.flatten()
euro_values = euro_values.flatten()
brazil_values = brazil_values.flatten()
world_values = world_values.flatten()
xticks = [x for x in range(1961, 2021)]
```

```python
plt.style.use('ggplot')
#make a plot 2X1

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), dpi=200)
axs[0].plot(xticks, usa_values, label='United States')
axs[0].plot(xticks, china_values, label='China')
axs[0].plot(xticks, euro_values, label='European Union')
axs[0].plot(xticks, brazil_values, label='Brazil')
axs[0].plot(xticks, world_values, label='World')
axs[0].xaxis.set_major_locator(MultipleLocator(10))
axs[0].xaxis.set_major_formatter('{x:.0f}')
axs[0].xaxis.set_minor_locator(MultipleLocator(5))
axs[0].yaxis.set_major_locator(MultipleLocator(1e9))
axs[0].yaxis.set_minor_locator(MultipleLocator(0.5e9))
axs[0].set_xlabel('Years')
axs[0].set_ylabel('population in billions')
axs[0].set_ylim(0.01e9, 8e9)
axs[0].set_xlim(1960, 2020)
axs[0].legend(loc= 2, labelcolor='black')

axs[1].plot(xticks, usa_values, label='United States')
axs[1].plot(xticks, china_values, label='China')
axs[1].plot(xticks, euro_values, label='European Union')
axs[1].plot(xticks, brazil_values, label='Brazil')
axs[1].xaxis.set_major_locator(MultipleLocator(10))
axs[1].xaxis.set_major_formatter('{x:.0f}')
axs[1].xaxis.set_minor_locator(MultipleLocator(5))
axs[1].yaxis.set_major_locator(MultipleLocator(1e9))
axs[1].yaxis.set_minor_locator(MultipleLocator(0.5e9))
axs[1].set_xlabel('Years')
axs[1].set_ylim(0.01e9, 1.5e9)
axs[1].set_xlim(1960, 2020)
axs[1].legend(loc= 2, labelcolor='black')

# plt.title('Total Population')
# plt.xlabel('Years')
# plt.ylabel('population in billions')
# plt.ylim(0.01e9, 8e9)
# plt.xlim(1960, 2020)
# plt.legend(loc= 2)
plt.show()
```
